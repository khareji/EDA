```python
delcol = [ 'id', 'imdb_id', 'popularity', 'budget_adj', 'revenue_adj', 'homepage', 'keywords', 
           'overview', 'production_companies', 'vote_count', 'vote_average'] #storing all unwanted columns

data = data.drop(delcol, 1) #droping unwanted columns

data.head(5)
```

```python
data.rename(columns = {'original_title' : 'movie_title', 'budget' : 'budget($)', 'revenue' : 'revenue($)'}, inplace = True) #renaming few coluns for our better understanding

data.head() # result after changing the name
```

```python
data.dtypes #checking again type of features
```

```python
data['release_date'] = data['release_date'].apply(pd.to_datetime) #chnaging date time format for better understanding
data.dtypes #previous it was object now it is in date time format
```

```python
data.insert(2, 'profit($)', data['revenue($)'] - data['budget($)']) #creating profit column .i.e profit=revenue -budget

data.head()
```

```python
highp=data["profit($)"].max() #max profit
highp
```

```python
lowp=data["profit($)"].min()#min profit
lowp
```

```python
highrevenue=data["revenue($)"].max() #max revenue
highrevenue
```

```python
lowrevenue=data["revenue($)"].min() #min revenue
lowrevenue
```

```python
highbudget=data["budget($)"].max() #max budget
highbudget
```

```python
lowbudget=data["budget($)"].min() #min budget
lowbudget
```

```python
avgruntime=data['runtime'].mean() #average runtime
avgruntime
```

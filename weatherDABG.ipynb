```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
delhi=pd.read_csv("delhi_data.csv")
allahabad=pd.read_csv("allahabad_data.csv")
banglore=pd.read_csv("bangalore_data.csv")
globaldata=pd.read_csv("global_data.csv")
```

```python
x=delhi["year"]
y=delhi["avg_temp"]
plt.plot(x,y,label='delhi')
plt.legend()
plt.xlabel("year")
plt.ylabel("temp in c")
plt.show()
```

```python
x=allahabad["year"]
y=allahabad["avg_temp"]
plt.plot(x,y,label='allahabad')
plt.legend()
plt.xlabel("year")
plt.ylabel("temp in c")
plt.show()
```

```python
x=banglore["year"]
y=banglore["avg_temp"]
plt.plot(x,y,label='banglore')
plt.legend()
plt.xlabel("year")
plt.ylabel("temp in c")
plt.show()
```

```python
x=globaldata["year"]
y=globaldata['avg_temp']
plt.plot(x,y,label='Global')
plt.xlabel("year")
plt.ylabel("temp in c")
plt.legend()
plt.show()
```
